```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import lxml
import json
import time
import csv
```

```python
# Записываем подгруженный файл excel в датафрейм, выводим информацию о нем

df = pd.read_excel('ID and links.xlsx')
```

```python
df
```

```python
# Формируем суп-объект

def get_soup(url):
  session = requests.session()
  session.headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36",
        "Accept-language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7"
  }

  res = session.get(url)
  html = res.text
  soup = BeautifulSoup(html, "lxml")
  return soup
```

```python
def char_parse(row):
  try:    
    soup = get_soup(row)
    table_pos = soup.select_one("td.right_info")
    table = table_pos.select("table.tableSort tbody tr")
    return table
  except:
    table = []
    return table
```

```python
df['char'] = df['Unnamed: 2'].apply(char_parse)
```

```python
# Записываем объедененные csv в итоговый Excel

writer = pd.ExcelWriter('dataset_parse1.xlsx') # write dataframe to excel 
df.to_excel(writer) # save the excel 
writer.save()
```
