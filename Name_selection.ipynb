```python
# Подключаем необходимые библиотеки 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import csv
import json
```

```python
# Записываем подгруженный файл excel в датафрейм, выводим информацию о нем

dp = pd.read_excel('prices (1).xlsx')
ds = pd.read_excel('sin_no_art.xlsx')
```

```python
dp = dp[['art', 'name']]
```

```python
dp
```

```python
# Делим короткое имя на части в список

ds['list'] = ds['sinonim'].apply(lambda x: x.split('-'))
ds
```

```python
print(type(ds['list'][2]))
ds['list'][2]
```

```python
shorts_list = list(dp['name'].values)
shorts_list
```

```python
# Генерим функцию проверки вхождения всех элементов короткого имени в длинное

def main(row):
  x = 0
  y = []
  for i in shorts_list:
    count_mod = 0
    for j in row:
      if j in i:
        count_mod += 1
      if count_mod == len(row):
        x += 1
        y.append(i)
  return [x, y]
```

```python
# Применяем функцию к серии list

ds['equal'] = ds['list'].apply(main)
ds
```

```python
# Выбираем строки с полным вхождением

def split_list(row):
  if row[0] > 0:
    l = row[1]
  else:
    l = []
  return l
```

```python
ds['small_name'] = ds['equal'].apply(split_list)
```

```python
df = ds[ds['small_name'].astype(bool)]
```

```python
writer = pd.ExcelWriter('cross_name1.xlsx') # write dataframe to excel 
df.to_excel(writer) # save the excel 
writer.save()
```
